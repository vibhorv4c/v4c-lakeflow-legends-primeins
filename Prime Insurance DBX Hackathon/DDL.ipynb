Create Primeins catalog

```python
%sql
CREATE CATALOG IF NOT EXISTS primeins;
```

Create schemas

```python
%sql
CREATE SCHEMA IF NOT EXISTS primeins.landing;
CREATE SCHEMA IF NOT EXISTS primeins.bronze;
CREATE SCHEMA IF NOT EXISTS primeins.silver;
CREATE SCHEMA IF NOT EXISTS primeins.gold;
```
